##  Akıllı Bildirim Sistemi (Smart Notification System)
Bu proje, Nesne Yönelimli Programlama (OOP) prensiplerini kullanarak farklı kanallar üzerinden bildirim gönderimini simüle eden modüler bir sistemdir.

1. Projenin Amacı
Sistemin temel amacı, tek bir yönetim arayüzü üzerinden farklı bildirim türlerini (Email, SMS, Push) kontrol etmektir. Bu süreçte Polymorphism (Çok Biçimlilik) ve Inheritance (Kalıtım) yapıları kullanılmıştır.

2. Polymorphism (Çok Biçimlilik) Nedir?
Farklı sınıflara ait nesnelerin, aynı isimdeki bir metodu (örneğin send()) çağırarak kendi sınıflarına özgü farklı davranışlar sergilemesidir. Bu sayede sistem, nesnenin türünden bağımsız olarak aynı komutu çalıştırabilir.

3. Kullanılan Python Özellikleri
  
Kalıtım: Base sınıfın özelliklerini alt sınıflara aktarması.

Özel Metodlar: __str__ ve __len__ gibi metodlarla nesne davranışlarının özelleştirilmesi.

Exception Handling: try-except blokları ile sistem güvenilirliğinin sağlanması.



```python
from notifications import EmailNotification, SMSNotification, PushNotification
from utils import generateMessage
```

```python
# Bildirim nesnelerini oluşturuyoruz
mail = EmailNotification(generateMessage())
sms = SMSNotification(generateMessage())
push = PushNotification(generateMessage())

# Bunları bir listeye koyuyoruz (Polymorphism için)
bildirimler = [mail, sms, push]

# Sonuçları ekrana yazdırıyoruz
print("--- Bildirim Sistemi Çalışıyor ---\n")

for b in bildirimler:
    print(b) # Bu __str__ metodunu tetikler
    b.send() # Bu polymorphism ile ilgili sınıfın send metodunu tetikler
    print("-" * 20)
```

```python
# Bildirimleri bir liste içinde topluyoruz
notifications = [mail, sms, push]

# Aynı kod satırı farklı çıktılar üretir
print("--- Polymorphism Çıktısı ---")
for notification in notifications:
    notification.send()
```

```python
# __str__() metodunun çalıştığını gösterelim
print(f"Nesne Yazdırılıyor: {mail}")

# __len__ ve normal len(message) farkını yorumlayalım
print(f"Mesajın gerçek uzunluğu (len): {len(mail.message)}")
print(f"Sınıf üzerinden __len__ sonucu: {len(mail)}")
```
